```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread(r'C:\Users\param\Pictures\Screenshots\Screenshot 2024-10-22 204844.png')

# Convert the image from BGR to RGB color space (OpenCV reads images in BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image_rgb, (15, 15), 0)
```

```python
# Convert the blurred image from RGB to BGR color space to save using OpenCV
blurred_image_bgr = cv2.cvtColor(blurred_image, cv2.COLOR_RGB2BGR)

# Save the blurred image
cv2.imwrite('blurred_image.jpg', blurred_image_bgr)
```

```python
# Display the images
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Display blurred image
plt.subplot(1, 2, 2)
plt.imshow(blurred_image)
plt.title('Blurred Image')
plt.axis('off')

plt.show()
```
